# Mix of Positional + Keyword

```python
def add_numbers(x,*,y): # accept anly keyvalue pair 
    return x + y
add_numbers(x=10,y=20)    
```

```python
Case 2 — Positional-only + normal parameters
def add_numbers(x, y, /, z):
    print(x, y, z)
add_numbers(1, 2, 3)        # ✅
add_numbers(1, 2, z=3)      # ✅
add_numbers(x=1, y=2, z=3)  # ❌ x,y must be positional
Rule:
x,y → positional only
z → flexible
```

# variable scope

```python
In Python, a variable's scope determines where it is visible and can be accessed or modified. 
When a variable is referenced, Python searches for it in the following order:
Local (L): The current function's scope.
Enclosing (E): The scope of any nested functions.
Global (G): The top-level scope of a module or script.
Built-in (B): The special scope for Python's pre-defined names, like print() and len(). 
```

```python
Local scope

A variable defined inside a function has a local scope.
It exists only for the duration of the function call and 
is not accessible from outside that function. 
Each call to the same function creates a new local scope. 
```

# editing global

```python
a=5
def f(): # not present in local but we try to change the global
    #a=5
    a+=1  # can not directly modify  may effect other program
    print(a)
f()    
```

```python
def f():
    global x  # now we can modify
    x+=1 
    print(x)
x=5
print(f())
print(x)
```

```python
now = datetime.datetime.now()
print(now)
formatted = now.strftime("%Y")
print("Formatted:", formatted)
```

```python
formatted = now.strftime("%b") # short month
print("Formatted:", formatted)
```

```python
formatted = now.strftime("%B") # full month
print("Formatted:", formatted)
```

```python
formatted = now.strftime("%d") # date/day
print("Formatted:", formatted)
```

```python
%Y → Year with century (2025)

%m → Month as number (01–12)

%B → Month name (November)

%d → Day of month (01–31)

%A → Weekday name (Friday)

%I → Hour (12-hour clock)

%H → Hour (24-hour clock)

%M → Minutes

%S → Seconds

%p → AM/PM
```

# date arithmetic

```python
from datetime import date, timedelta

today = date.today()

tomorrow = today + timedelta(days=-365)
yesterday = today +  timedelta(days=-1)
print(today)
print("Tomorrow:", tomorrow)
print("Yesterday:", yesterday)
```

```python
#Difference between two dates
diff=tomorrow -yesterday
print(diff)
print(type(diff))
```
