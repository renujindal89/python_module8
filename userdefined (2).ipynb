# variable

```python
In Python, a variable's scope determines where it is visible and can be accessed or modified. 
When a variable is referenced, Python searches for it in the following order:
Local (L): The current function's scope.
Enclosing (E): The scope of any nested functions.
Global (G): The top-level scope of a module or script.
Built-in (B): The special scope for Python's pre-defined names, like print() and len(). 
```

```python
Local scope

A variable defined inside a function has a local scope.
It exists only for the duration of the function call and 
is not accessible from outside that function. 
Each call to the same function creates a new local scope. 
```

```python
def my_function():
    x = 10  # `x` has a local scope
    print(x) # This will print 10
   

my_function()
# Output: 10

#print(x)
# Output: NameError: name 'x' is not defined
#print(x)
```

```python
y=20    #global scope
def my_function():
    x = 10  # `x` has a local scope
    return x # This will print 10

print(my_function())
z=10
# Output: 10
print(y)
print(z)
## Output: NameError: name 'x' is not defined
```

```python
# local and global can have a same name 
# both variable scope are diff
x=20
y=30
def f():
    x = 10  # `x` has a local scope
    return x # This will print 10

print(f())

# Output: 10
print(x)
```

# enclosing scope

```python
In programming—especially in Python—an enclosing scope variable refers to a variable that:

Is defined in an outer (enclosing) function, and

Is used inside an inner (nested) function.
```

```python
def outer():
    x = 10   # <-- enclosing scope variable

    def inner():
        return x  # inner function uses variable from the enclosing scope

    return inner

f = outer()
print(f())   # prints 10
```

# Global scope

```python
A variable defined at the top level of a script or module has a global scope
and is accessible throughout the program
```

```python
z = 40  # `z` has a global scope

def my_function():
    print(z)  # Can access `z` from the global scopeprint(z)
my_function()
print(z)
```

# wants to modify the global  variable

```python
z = 40  # global

def my_function():
    z += 1    # tries to modify z
    print(z)

my_function()
```

```python
z = 40  # `z` has a global scope

def my_function():
    global z  # tell Python to use the global z
    z += 1  # tries to modify z
    print(z)
my_function()
print(z)
```

```python
Using nonlocal (for enclosing function variables)

nonlocal lets you modify a variable from the enclosing (non-global) scope —
meaning a variable defined in an outer function but not global.
```

```python
def outer():
    x = 10   # enclosing scope variable

    def inner():
        nonlocal x   # use x from outer()
        x += 1
        print("inner:", x)

    inner()
    print("outer:", x)

outer()
```

# Built-in scope

```python
This is the broadest scope, containing all of Python's pre-defined names,
like built-in functions (print, len) and exceptions. 

This scope is loaded when the Python interpreter starts and is always available. 

It is generally bad practice to override or "shadow" these built-in names with your own variables.  
```

```python
import builtins
print(dir(builtins))
```

```python
#Here, len (which is normally a built-in function that returns the length of objects)
is now assigned the integer value 5.

So from this point onward, inside your current scope,
len no longer refers to the function — it refers to the number 5.
```
