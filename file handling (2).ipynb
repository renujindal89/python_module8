```python
# write multiple line
#Writes a list( or any iterable) of strings to the file.
#Does not automatically add \n.
#You must include newline characters manually.
l=['hello\n','hi\n','how are you\n','I am fine\n']
f=open('sample4.txt','a')
f.writelines(l)
f.close()
```

```python
When to use what?
✔ Use write() when:

Writing one line or one string at a time

Building file content manually

✔ Use writelines() when:

You already have a list of strings

You want to write many lines at once
```

# SEEK AND TELL FUNCTION

```python
#file.tell() returns the current position of the file pointer
#“Where am I currently inside the file?”
with open(r'C:\Users\complere\Desktop\demo\sql1.txt','r') as f:
    print(f.read(150))
    print(f.tell())
    print(f.read(50))
    print(f.tell())
```

```python
# f.seek() is used to move the file pointer to a specific position in the file.
with open(r'C:\Users\complere\Desktop\demo\sql1.txt','r') as f:
    print(f.read(100))
    print(f.tell())
    f.seek(90)
    print(f.read(5))
    print(f.tell())

```

# csv--->list

```python
#create a csv file using list
list--->csv   comma seperated file
csv--->list
```

```python
import csv

# Data to write (list of lists or list of tuples)
data = [
    ['name', 'age', 'city'],  # header row
    ['Alice', 30, 'New York'],
    ['Bob', 25, 'Los Angeles'],
    ['Charlie', 35, 'Chicago']
]

# Open a new file in write mode
with open('people10.csv', 'w') as f:
    writer = csv.writer(f)

    # Write all rows
    writer.writerows(data)
```

# read a csv file  

```python
#rows as lists)
import csv

with open('people10.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)  # each row is a list of strings
# shows empty lists [] after every row.
#This usually happens because the CSV file has extra blank lines between the rows.        
```

```python
#rows as dictionaries
import csv

with open('people10.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)  # each row is a dict with column names as keys
```

```python
Using JSON (BEST method for lists/dicts)

If you are storing Python lists or dictionaries, JSON is the safest and easiest way.
```

```python
#Write list or dict to a file
import json

data = {
    "name": "Alice",
    "age": 21,
    "marks": [88, 92, 79]
}

with open("data10.json", "w") as f:
    json.dump(data, f)  # {"name": "Alice", "age": 21, "marks": [88, 92, 79]}
```

```python
#Read list/dict from a file
import json

with open("data3.json", "r") as f:
    data = json.load(f)

print(data)
```
