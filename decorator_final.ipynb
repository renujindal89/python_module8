# Basic decorator concept

```python
def decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper
```

```python
A decorator is just a function
The @ syntax is special Python syntax to apply it
wrapper is just a function inside another function
It “wraps” the original function → that’s why the name
```

# Using decorator manually

```python
def say_hi():
    print("Hi")

say_hi = decorator(say_hi)
say_hi()
```

# with argument

```python
def decorator(func):
    def wrapper(a,b):
        print("Start of function")
        result = func(a,b)  # call original function and save result
        print("End of transaction")
        return result  # return the result to the caller
    return wrapper

# Example function
```

```python
def add(x, y):
    print("Inside function")
    return x + y

add = decorator(add)
print(add(2,3))
```

```python
@decorator
def add(x, y):
    print("Inside function")
    return x + y
add(2,3)    
```

```python
# #Write a decorator that checks if a number is positive before executing the function.
```

```python
def check_positive(func):
    def wrapper(n):
        if n <= 0:
            print("Number must be positive")
            return
        return func(n)
    return wrapper
```

```python
@check_positive
def square(n):
    print("Inside function")
    return n * n

print(square(5))    # valid
print(square(-3))   # invalid
```
