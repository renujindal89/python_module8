```python
What Is a Regular Expression (Regex)?

A Regex is a mini-language used to search, match, and extract patterns inside text.

It helps you find things like:

Only letters

Only numbers

Email patterns

Dates

Words starting with “A”

Anything that matches a rule
```

```python
SELECT * FROM users
WHERE name = 'Renu';


SELECT * FROM users
WHERE name LIKE 'R%';


✔ Easy
❌ Limited (no complex patterns)
```

# 🔹 2. REGEXP (Regular Expression)

```python
SELECT * FROM users
WHERE name REGEXP '^R';

👉 ^R → starts with R

2.
✔ Ends with
WHERE name REGEXP 'n$'


3.✔ Contains digits
WHERE name REGEXP '[0-9]'

4.✔ Only alphabets
WHERE name REGEXP '^[A-Za-z]+$'  # + means one or more occurrences
```

```python
-- LIKE
WHERE email LIKE '%@gmail.com'

-- REGEXP
WHERE email REGEXP '^[A-Za-z0-9._%+-]+@gmail\\.com$'
```

# Example in Python

```python
re.match()

👉 Checks only at the beginning of the string
```

```python
import re

text = "123abc"
print(re.match(r"\d+", text))
```

```python
print(re.match(r"[a-z]+", "123abc"))  # ❌ None
```

```python
re.fullmatch()

👉 Checks entire string must match
```

```python
re.fullmatch(r"\d+", "123")    # ✅ Match
re.fullmatch(r"\d+", "123abc") # ❌ None
```

```python
import re

pattern = r'[A-Za-z]+'

text = "Renu123"

if re.fullmatch(pattern, text):
    print("Valid")
else:
    print("Invalid")
```

```python
import re

tests = ["Renu", "abcXYZ", "abc123", "", "Hello World"]

for t in tests:
    if re.fullmatch(r'[A-Za-z]+', t):
        print(t, "→ Valid")
    else:
        print(t, "→ Invalid")
```

```python
# re.findall()
 Finds all matches anywhere in string
```

```python
text = "abc123xyz45"

print(re.findall(r"\d+", text))
```

```python
import re

text = "Order123Number45"

print(re.findall(r"\d", text))   # single digits
print(re.findall(r"\d+", text))  # full numbers
```

```python
[A-Za-z]	   Any letter	              A, d, Z
[0-9]	        Any digit	                    3
[A-Za-z0-9]  	Letters or digits	      A, 9
[^A-Za-z]	    NOT a letter	         1, -, @
```

```python
\d	one digit (0–9)
\d+	one or more digits
\d{4}   exactly 4 digit
```

```python
^	Start of string
$	End of string
{n}Exactly n
{2,} atleast 2 
{2,5} → between 2 and 5 times
```
