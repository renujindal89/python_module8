# constructor

```python
A constructor in Python is a special method that runs automatically when you create an object.
```

```python
class Student:
    def __init__(self, name, age):
        
        self.name = name
        self.age = age
        
        print(self.name, self.age)
        

    def show(self):
        print(self.name, self.age)

# object creation
s1 = Student("Renu", 21)
s1.show()
```

```python
class Student:
    def __init__(self, name, age):   # constructor
        self.name = name
        self.age = age

s1 = Student("Renu", 21)
print(s1.name, s1.age)
```

# Encapsulation

```python
“Hide internal data and allow access only through methods.”
```

```python
Encapsulation = wrapping data (variables) + methods together and controlling access to them
```

```python
Why use Encapsulation?
Protect data from accidental changes
Control how data is modified
Make code secure and maintainable
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name        # public
        self._marks = marks     # protected (by convention)
        self.__grade = "A"      # private

    def display(self):
        print(self.name, self._marks, self.__grade)

s = Student("Renu", 90)
s.display()
```

```python
s.name
```

```python
Access types in Python (correct understanding)
✅ 1. Public
self.name = "Renu"

✔ Access from anywhere
✔ Update from anywhere

s.name = "Amit"   # ✅ allowed
```

```python
⚠️ 2. Protected (_var)
self._marks = 90

✔ Can access outside
✔ Can update outside
❗ But should NOT (by convention)

s._marks = 50   # ⚠️ allowed but not recommended

👉 Means: “internal use, please don’t touch”
```

```python
🔒 3. Private (__var)
self.__balance = 1000

❌ Cannot access directly outside
❌ Cannot update directly outside
✔ Access/update via methods

print(s.__balance)   # ❌ Error
```

```python
class Bank:
    def __init__(self):
        self.__balance = 1000     #private

    def deposit(self, amount):
        self.__balance += amount   # ✅ update allowed

b = Bank()
b.deposit(500)
```

# inheritance

```python
Inheritance = one class (child) gets properties and methods of another class (parent).
```

```python
class Person:              # Parent class
    def show(self):
        print("I am a person")

class Student(Person):     # Child class (inherits Person)
    def study(self):
        print("I study")

s = Student()
s.show()    # inherited method
s.study()   # own method
```

# Why use Inheritance?

```python
Reuse existing code
Avoid duplication
Extend functionality
```

```python
# single level -> parent->child
class car():
    def start(self):
        print(" car started")
    def stop(self):
        print(" car stopped")  
class tyotacar():
    def __init__(self,name):
        self.name=name
car1=tyotacar("fortuner")  
car1.start()
car1.stop()
car1.name       
```

```python
# single level -> parent->child
class car():
    def start(self):
        print(" car started")
    def stop(self):
        print(" car stopped")  
class tyotacar(car):
    def __init__(self,name):
        self.name=name
car1=tyotacar("fortuner")  
car1.start()
car1.stop()
car1.name                       # sum()
```

```python
# multiple inheritance  #parent-------->parent----------> child
class a():
    var1="welcome to class a"
class b():
    var2="welcome to class b"
class c(a,b):
    var3="welcome to class c"
c1=c()
print(c1.var3)
print(c1.var2)
print(c1.var1)
```

# Operator Overloading

```python
# polymorphism
print(1+2) #3
print("renu"+"jindal") # concatination
print([1,2,3]+[4,5,6]) # merge
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    # Overloading the + operator
    def __add__(self, other):
        return self.value + other.value

a = Number("renu")
b = Number("jindal")
print(a + b)  # 15
```

```python
a = Number(10)
b = Number(5)
print(a + b)  # 15
```

```python
a = Number([1,2,3])
b = Number([5,6,7])
print(a + b)  # 15
```
