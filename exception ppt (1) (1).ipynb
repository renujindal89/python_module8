# try-except

```python
a=10
b=0           
try:                                                          #in try we may say this is a critical statement i am not sure ABOUT THIS it try to execute        
    print(a/b)
except Exception:                                                       #except says i will accept the error which will come in try.  
    print("Hey, you cannot divide a number by zero")

print("Idea")
```

```python
Catch specific exceptions (best practice)
try:
    x = int(input("Enter number: "))
    print(10 / x)
except ValueError:
    print("Invalid input! Please enter a number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
```

```python
# Using as to get error message
```

```python
try:
    x = int("abc")
except ValueError as e:
    print("Error:", e)
```

```python
a=10
b=0         
try:                                                            
    print(a/b)
except Exception as e:                                                      
    print("Hey, you cannot divide a number by zero", e)

```

```python
#Using else (runs if no exception)
try:
    x = int("10")
except ValueError:
    print("Error")
else:
    print("Success:", x)
```

```python
# Using finally (always runs)
```

```python
try:
    f = open("file.txt")
except FileNotFoundError:
    print("File not found")
finally:
    print("This will always execute")
```

```python
try:
    # risky code
    pass
except SomeError:
    pass
else:
    pass
finally:
    pass
```

```python
#try except finally
a=10
b=0         
try: 
    print("resource open")
    print(a/b)
except Exception as e:                                                      
    print("Hey, you cannot divide a number by zero", e)
finally:                                              #it will always execute either exception is there or not
    print("resource close")
```

```python
#special case
a=10
b=0     


try: 
    print("resource open")
    print(a/b)
    k = int(input("enter a number"))
    print(k)

except ZeroDivisionError as e:                                                      
    print("Hey, you cannot divide a number by zero", e)

except ValueError as v:                                                      
    print("enter correct value",v)

except Exception:
    print("something went wrong....")
    
finally:                                            
    print("resource close")
```
